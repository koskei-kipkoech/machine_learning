**MatplotLib**

MatplotLib is a plotting Library for the Python Programming Language and it's numerical mathematics extension Numpy. It Provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, WxPython, Qt or Gtk+

**Pros of MatplotLib are:**
*   Generally easy to get Started for simpler plots
*   Support for custom lebels and texts
* Greatcontrol of every element in a figure.
* High-quality output in many formats.
* Very customizable in general




```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
##Plotting using matplotLib

## plit scatter
plt.scatter(x,y,c='teal')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot Graph')
plt.show()
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*--',linestyle='dotted', linewidth=2, markersize=10)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot Graph')
plt.show()
```

```python
## Creating a Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo-')
```

```python
x= np.arange(1,11)
y= 3 * x + 5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y,label="my line")
#plt.legend()
plt.show()
```

```python
#compute the x and y coordinates for points on a curve sine curve

x=np.arange(0,4 * np.pi, 0.1)
y=np.sin(x)
plt.title("Sine curve")

#Plot the points using matplotLib
plt.plot(x,y, "r--")
plt.show()
```

```python
#Subplot()
#Combine the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#setup a subplot grid that has height 2 and width 1,
# and set the first such subplot as active
plt.subplot(2, 1, 1)

#make the first plot
plt.plot(x, y_sin,'g--')
plt.title('Sine')

#set the second subplot as active and make the second plot
plt.subplot(2, 1, 2)
plt.plot(x, y_cos,'r--')
plt.title('Cosine')

#show the figure
plt.show()
```

**Bar Graph**

```python
# Bar plot
x = [2,8,10]
y = [11,16,9]
x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x, y, align='center')
plt.bar(x2, y2, color='g', align='center')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

**Histogram**

```python
a= np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins=20)
plt.title("histogram")
plt.show()
```

**Box Plot using MatplotLib**


```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# Retangular box plot
plt.boxplot(data,
            vert=True,  # vertical box alignment
            patch_artist=True,  # fill with color
            # fill with colors
            labels=['x1', 'x2', 'x3'])  # will be used to label x-ticks
```

**Pie Chart**

```python
#Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0, 0.1, 0)  # explode 1st slice

#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
